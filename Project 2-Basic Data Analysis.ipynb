```python
#####Basic Data Analysis with Python
#
# First !pip install pandas and libraries are needed pandas for data maniuplation and matplotlib for visualization of the data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# I downloded the data from kaggle about the population happiness
happiness_data = pd.read_csv('/Users/behnam/Downloads/2019.csv')
```

```python
#display the first few rows of the dataset
print(happiness_data.head())
```

```python
#display basic info
print(happiness_data.info())
```

```python
#check for missing data in each column or drop or forward fill the missing value
# Example: happiness_data.dropna(inplace=True)
# or
# happiness_data.fillna(method='ffill', inplace=True)
print(happiness_data.isnull().sum())
```

```python
#for statical data like mean, percentiles
print(happiness_data.describe())
```

```python
#Grouping by region and calculating the mean
region_means = happiness_data.groupby('Country or region')['Score'].mean()
print(region_means)
```

```python
#creating a bar plot for creating score of average happiness for every region
plt.figure(figsize=(15,6))
region_means.plot(kind='bar')
plt.title('Av HAP PER REG')
plt.xlabel('Region')
plt.ylabel('Score')
plt.xticks(rotation=90, fontsize=6)
plt.show()
```

```python
plt.scatter(happiness_data['GDP per capita'], happiness_data['Score'])
plt.title('GDP per capita Vs. H-Score')
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.show()
```
